# Разложением методом STL на тренд и сезонность 

```python
import numpy as np
import pandas as pd

from sktime.transformations.series.detrend import STLTransformer

from sktime.utils.plotting import plot_series

import pandas_datareader.data as web

# настройки визуализации
import matplotlib.pyplot as plt

# Не показывать Warnings
import warnings
warnings.simplefilter(action='ignore', category=Warning)
```

Загрузим из БД [`FRED`](https://fred.stlouisfed.org/) дневные данные по Gross Domestic Product (Symbol [`NA000334Q`](https://fred.stlouisfed.org/series/NA000334Q)) с 1990-01-01 по 2025-12-31 (ряд `z`) и создадим датафрейм `y=log(z)`

```python
z = web.DataReader(name='NA000334Q', data_source='fred', start='1990-01-01',  end='2025-12-31')
y = np.log(z)
```

## Выделение тренда

```python
transformer = STLTransformer(trend_deg=1, sp=4, return_components=False)
y_trend = transformer.fit_transform(y)

plot_series(y, y_trend)
```

## Разложением ряда на компоненты

__Замечание__: в итоговом разложении `transformed=trend+resid`

```python
transformer = STLTransformer(trend_deg=1, sp=4, return_components=True)
decomposition = transformer.fit_transform(y)
decomposition
```

```python
decomposition.plot(subplots=True)
```
