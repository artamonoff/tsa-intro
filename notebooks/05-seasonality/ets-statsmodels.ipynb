# Модель ETS (пакет [`statsmodels`](https://www.statsmodels.org/stable/index.html))

```python
import numpy as np
import pandas as pd

from statsmodels.tsa.api import ETSModel

import pandas_datareader.data as web

# настройки визуализации
import matplotlib.pyplot as plt

# Не показывать Warnings
import warnings
warnings.simplefilter(action='ignore', category=Warning)
# Не показывать ValueWarning, ConvergenceWarning из statsmodels
from statsmodels.tools.sm_exceptions import ValueWarning, ConvergenceWarning
warnings.simplefilter('ignore', category=ValueWarning)
warnings.simplefilter('ignore', category=ConvergenceWarning)
```

Загрузим из БД [`FRED`](https://fred.stlouisfed.org/) дневные данные по Gross Domestic Product () (Symbol [`NA000334Q`](https://fred.stlouisfed.org/series/NA000334Q)) с 1990-01-01 по 2025-12-31 (ряд `z`) и создадим датафрейм `y=log(z)`

```python
z = web.DataReader(name='NA000334Q', data_source='fred', start='1990-01-01', end='2025-12-31')
y = np.log(z)

y.plot()
plt.show()
```

### Аддитивная модель для `y`

Подгоним модель ETS(A,A,A) и построим прогноз. Число сезонов выберем равным 4

```python
# спецификация модели
mod = ETSModel(y.squeeze(), error='add', trend='add', seasonal='add', seasonal_periods=4, damped_trend=False)
# подгонка модели на данных
res = mod.fit()
# выводим результаты подгонки
res.summary(alpha=0.05)
```

```python
# прогнозирование
forecasts = res.forecast(steps=10)
forecasts
```

Визуализация прогноза

```python
plt.plot(y.tail(30), label='y')
plt.plot(forecasts, label='y_pred')
plt.legend()

plt.show()
```

### Аддитивная модель для `y` с демпфированным трендом

Подгоним модель ETS(A,A,A) и построим прогноз. Число сезонов выберем равным 4

```python
# спецификация модели
mod = ETSModel(y.squeeze(), error='add', trend='add', seasonal='add', seasonal_periods=4, damped_trend=True)
# подгонка модели на данных
res = mod.fit()
# выводим результаты подгонки
res.summary(alpha=0.05)
```

```python
plt.plot(y.tail(30), label='y')
plt.plot(forecasts, label='y_pred')
plt.legend()

plt.show()
```

### Мультипликативная модель для `z`

```python
# спецификация модели
mod = ETSModel(z.squeeze(), error='mul', trend='mul', seasonal='mul', seasonal_periods=4, damped_trend=False)
# подгонка модели на данных
res = mod.fit()
# выводим результаты подгонки
res.summary(alpha=0.05)
```

```python
plt.plot(z.tail(30), label='z')
plt.plot(forecasts, label='z_pred')
plt.legend()

plt.show()
```
