```python
import numpy as np
import pandas as pd

from sktime.transformations.series.hpfilter import HPFilter

from sktime.utils.plotting import plot_series

import pandas_datareader.data as web

# настройки визуализация
import matplotlib.pyplot as plt

# Не показывать Warnings
import warnings
warnings.simplefilter(action='ignore', category=Warning)
```

Загрузим из БД [`FRED`](https://fred.stlouisfed.org/) дневные данные по Gross Domestic Product (Symbol [`NA000334Q`](https://fred.stlouisfed.org/series/NA000334Q)) с 1990-01-01 по 2025-12-31 (ряд `z`) и создадим датафрейм `y=log(z)`

```python
z = web.DataReader(name='NA000334Q', data_source='fred', start='1990-01-01')
y = np.log(z)
```

```python
hp = HPFilter(1600)
y_cycle = hp.fit_transform(y)
y_trend = y - y_cycle
```

```python
plot_series(y_cycle)
plt.show()
```

```python
plot_series(y, y_trend, labels=['level', 'trend'])
plt.show()
```
